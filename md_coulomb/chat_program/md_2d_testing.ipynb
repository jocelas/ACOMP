```python
from md_2d import *
from IPython.display import HTML
import time
```

```python
potential = lambda r: 100./r

system = MDSimulation2D(100, 4, 1e-3, 10000, init_positions='random', initial_temperature=0, target_temperature=.1, ensemble = 'NVT', boundary='periodic', potential_func=potential)
#system.plot_temperature()
#system.plot_energies()
```

```python
system.reset()
rho = system.density
T = system.initial_temperature

i = time.time()

system.animate_trajectories(interval = 10, steps_per_frame=2, show_vectors=True, vector_scale=0.01, frames=700, save_path=f"{i}_1temp_grid_rho_{rho}.mp4")
```

```python
system.reset()
system.run()
system.plot_energies()
system.plot_temperature()
```

```python
kin_e = system.kinetic_energy_history
times = np.arange(len(kin_e))
plt.plot(times, kin_e)
system.plot_pressure()
```
